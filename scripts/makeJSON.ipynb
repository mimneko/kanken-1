```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    # PDFを開く
    document = fitz.open(pdf_path)

    # ページごとにテキストを抽出
    for page_num in range(document.page_count):
        page = document.load_page(page_num)
        text = page.get_text("text")
        print(f"Page {page_num + 1}:\n{text}\n")

if __name__ == "__main__":
    # PDFファイルのパスを指定
    pdf_path = "onyomi.pdf"
    extract_text_from_pdf(pdf_path)
```

```python
import pandas as pd

def extract_highlights(old_csv_path, new_csv_path):
    df = pd.read_csv(old_csv_path)

    # ハイライトの前、ハイライト、ハイライトの後を抽出
    df[['前部分', 'ハイライト', '後部分']] = df['文章'].str.extract(r'(.*?)【(.*?)】(.*)')

    # 元のCSVを上書き
    df.to_csv(new_csv_path, index=False)

if __name__ == "__main__":
    extract_highlights("onyomi.csv", "onyomi+.csv")
    extract_highlights("kunyomi.csv", "kunyomi+.csv")
```

```python
import pandas as pd

def check_highlight_empty(csv_path):
    # CSVファイルを読み込む
    df = pd.read_csv(csv_path)

    # ハイライト列が空の行を探す
    empty_highlight_rows = df[df['ハイライト'].isna()]

    # 空の行があるかどうかをチェック
    if not empty_highlight_rows.empty:
        print("ハイライト列に値が存在しない行があります。")
        # 行番号を表示
        for idx in empty_highlight_rows.index:
            print(f"行番号: {idx}")
    else:
        print("すべての行にハイライトがあります。")

if __name__ == "__main__":
    check_highlight_empty("onyomi+.csv")
    check_highlight_empty("kunyomi+.csv")
```

```python
def add_answers(highlight_csv, answers_csv, output_csv):
    # ハイライトCSVと解答CSVを読み込む
    highlight_df = pd.read_csv(highlight_csv)
    answers_df = pd.read_csv(answers_csv)

    # マージする
    merged_df = pd.merge(highlight_df, answers_df, how='left', left_on='ハイライト', right_on='問題')

    # 必要な列を出力用に選択
    output_df = merged_df[['前部分', 'ハイライト', '後部分', '解答', '別答']]

    # 新しいCSVとして保存
    output_df.to_csv(output_csv, index=False)

if __name__ == "__main__":
    # "onyomi+.csv" と "A1音読み.csv" をマージして "onyomi++.csv" を作成
    add_answers("onyomi+.csv", "A1音読み.csv", "onyomi++.csv")
```

```python
# prompt: onyomi+.csvのハイライトの要素、またA1音読み.csvの問題の要素、の順番が同じことを確認せよ

def check_order(highlight_csv, answers_csv):
    # ハイライトCSVと解答CSVを読み込む
    highlight_df = pd.read_csv(highlight_csv)
    answers_df = pd.read_csv(answers_csv)

    # ハイライト列と問題列をリストとして抽出
    highlight_list = highlight_df['ハイライト'].tolist()
    answers_list = answers_df['問題'].tolist()

    # 要素の順番が同じかどうかをチェック
    if highlight_list == answers_list:
        print("onyomi+.csvのハイライトとA1音読み.csvの問題の順番は同じです。")
    else:
        print("onyomi+.csvのハイライトとA1音読み.csvの問題の順番は異なります。")

if __name__ == "__main__":
    check_order("onyomi+.csv", "A1音読み.csv")
```

```python
import pandas as pd
import re

def process_text(text):
    if '≒' not in text:
        return text

    parts = text.split('≒')

    processed_parts = []
    start_index = 1 if text.startswith('≒') else 0

    for part in parts[start_index:]:
        match = re.search(r'[┌├└〈]', part)
        if match:
            processed_parts.append(part[match.start():])
        else:
            processed_parts.append(part)

    return ''.join(processed_parts)

# DataFrameの各行に適用
new_df['文章'] = new_df['文章'].apply(process_text)

if __name__ == "__main__":
    df = pd.read_csv("A1音読み.csv")

    # 問題列から熟語を抽出
    df['熟語'] = df['問題'].str.extract(r'【(.*?)】')
    new_df = df[['ID', '難易度', '熟語', '意味']].copy()

    new_df['読み'] = df['解答'].replace('\n', '、')

    # df['別答'] に値がある行をフィルタリング
    df_with_answers = df[df['別答'].notna()]

    # フィルタリングされた行に対して処理を実行
    for index, row in df_with_answers.iterrows():
        # 同じインデックスを持つ new_df の行にアクセス
        new_df.loc[index, '読み'] = new_df.loc[index, '読み'] + "、" + row['別答'].replace('\n', '、')

    # 列をコピー
    new_df['文章'] = new_df['意味']

    # 文章列から不要な部分を削除
    new_df['文章'] = new_df['文章'].str.replace(r'★.*', '', regex=True)
    new_df['文章'] = new_df['文章'].str.replace(r'◇.*', '', regex=True)
    new_df['文章'] = new_df['文章'].str.replace(r'●.*', '', regex=True)
    #new_df['文章'] = new_df['文章'].str.replace(r'≒.*?┌', '┌', regex=True)
    #new_df['文章'] = new_df['文章'].str.replace(r'≒.*?├', '├', regex=True)
    #new_df['文章'] = new_df['文章'].str.replace(r'≒.*?└', '└', regex=True)
    #new_df['文章'] = new_df['文章'].str.replace(r'≒.*?〈', '〈', regex=True)
    new_df['文章'] = new_df['文章'].apply(process_text)
    new_df['文章'] = new_df['文章'].str.split('〈', n=1, expand=True)[0]
    new_df['文章'] = new_df['文章'].str.replace(r'┌', '')
    new_df['文章'] = new_df['文章'].str.replace(r'├', '')
    new_df['文章'] = new_df['文章'].str.replace(r'└', '')
    new_df['文章'] = new_df['文章'].str.replace('\n', '')
    new_df['文章'] = new_df['文章'].str.replace(r'\(.*?\)', '', regex=True)

    # 番号を削除
    while True:
        previous_df = new_df.copy()  # 変更前の状態を保存
        new_df['文章'] = new_df['文章'].str.replace(r'\d+\.', '', regex=True)  # 数字とピリオドを削除

        if new_df['文章'].equals(previous_df['文章']):  # 変更がなければループ終了
            break

    # 空白行を削除
    new_df = new_df[new_df['文章'].str.strip() != '']

    # 意味
    new_df['意味'] = new_df['熟語'] + '（' + new_df['読み'] + '）\n' + new_df['意味']
    new_df['意味'] = new_df['意味'].str.replace(r'◇.*', '', regex=True)
    new_df['意味'] = new_df['意味'].str.replace(r'●.*', '', regex=True)
    new_df['意味'] = new_df['意味'].str.rstrip('\n')

    # 新しいCSVとして保存
    print(new_df)
    new_df.to_csv("A3意味.csv", index=False, encoding='utf-8-sig')
```

```python
# prompt: E4四字熟語意味.csvを開き、語群列の「・」を改行に変換
# すべての行において、語群を改行コードで分割したものの中から、意味列の先頭４文字をそれぞれの文字を１文字ずつみて、すべて含むもののみを抽出し、四字熟語列とする。当てはまるものがなければ、IDとともにprintで吐くように
# E4四字熟語意味.csvにbomつきutfで保存

# E4四字熟語意味.csvを開き、語群列の「・」を改行に変換
df = pd.read_csv("E4四字熟語意味.csv")
df['語群'] = df['語群'].str.replace('・', '\n')

# 四字熟語列を追加
df['四字熟語'] = ''

# 各行に対して処理
for index, row in df.iterrows():
    meaning_start = row['意味'][:4]  # 意味の先頭4文字
    candidates = row['語群'].split('\n')  # 語群を改行で分割

    found = False
    for candidate in candidates:
        if all(char in candidate for char in meaning_start):  # 4文字すべてが含まれているかチェック
            df.loc[index, '四字熟語'] = candidate
            found = True
            break

    if not found:
        print(f"ID {row['ID']} に該当する四字熟語が見つかりません。")

df['ハイライト'] = df['四字熟語'].str.extract(r'\[(.*?)\]')
df['四字熟語'] = df['四字熟語'].str.replace('[', '').str.replace(']', '')

# E4四字熟語意味.csvにbomつきutfで保存
df.to_csv("E4四字熟語意味.csv", index=False, encoding='utf-8-sig')
```

```python
# prompt: D4四字熟語書き取り.csvを読み込み、各行ごとに「別読み」列を改行コードで分割し、「読み」列と同じものを除いたあと、そのリストを改行コードで接続し「別読み」列を上書きする。読み列と同じものがなければ、表示

import pandas as pd

df = pd.read_csv("D4四字熟語書き取り.csv")

# 各行に対して処理
for index, row in df.iterrows():
    if pd.notna(row['別読み']):  # 別読み列が空でない場合のみ処理
        alternative_readings = row['別読み'].split('\n')  # 改行コードで分割
        filtered_readings = [reading for reading in alternative_readings if reading != row['読み']]  # 読み列と同じものを除外
        if filtered_readings:  # 除外後も要素が残っている場合
            df.loc[index, '別読み'] = '\n'.join(filtered_readings)  # 改行コードで接続して上書き
        else:
            print(f"ID {row['ID']} には読み列と異なる別読みが存在しません。")

# 結果を表示または保存
print(df)
# df.to_csv("D4四字熟語書き取り_processed.csv", index=False)  # 必要であれば保存
```

```python
# @title A1音読み.csv -> 1-1.json
import pandas as pd
import json

# CSVファイルを読み込む
df = pd.read_csv("A1音読み.csv")

# JSONデータの初期構造
json_data = {
    "type": "1-1",
    "write_date": get_today_date(),
    "title": "音読み",
    "subQuestions": []
}

df['問題'] = df['問題'].apply(lambda x: x + '。' if not x.endswith('。') else x)

# 各行を処理してsubQuestionsに追加
for _, row in df.iterrows():
    # 問題部分
    if '【' in row['問題'] and '】' in row['問題']:
        before_part, rest = row['問題'].split('【', 1)
        highlight_part, after_part = rest.split('】', 1)
        question_parts = {
            "before": before_part,
            "highlight": highlight_part,
            "after": after_part
        }
    else:
        print(f"問題に【】がない\n{row['問題']}")

    # 解答部分 (別答がある場合は結合)
    answer = row['解答'].replace('\n', '、')

    # 意味部分 (●以降と最後の改行を削除)
    meaning = row['意味']
    if '◇' in meaning:
        meaning = meaning.split('◇')[0]
    if '●' in meaning:
        meaning = meaning.split('●')[0]
    if '＊' in meaning:
        meaning = meaning.split('＊')[0]
    meaning = meaning.replace('┌', '').replace('├', '').replace('└', '')
    meaning = meaning.rstrip('\n').replace('\n', '<br>').replace('👉', '→')

    # ID
    id_value = row['ID']

    # 難易度
    level_value = int(row['難易度'].replace('◆', ''))

    # サブ質問データを作成
    sub_question = {
        "id": id_value,
        "level": level_value,
        "questionParts": question_parts,
        "answer": answer,
        "explanation": meaning
    }

    # サブ質問をJSONデータに追加
    json_data["subQuestions"].append(sub_question)

# JSONファイルに書き込む
with open("1-1.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=4)
```

```python
# @title H7対義語類義語.csv -> 7-1.json
import pandas as pd
import json
import jaconv

# CSVファイルを読み込む
df = pd.read_csv("H7対義語類義語.csv")

# 「指定」が「⇔」のものだけを抽出
df = df[df['指定'] == '⇔']

# JSONデータの初期構造
json_data = {
    "type": "7-1",
    "write_date": get_today_date(),
    "title": "対義語",
    "subQuestions": []
}

# 各行を処理してsubQuestionsに追加
for _, row in df.iterrows():
    # 問題部分
    question_parts = {"before": row['問題'].replace('【', '').replace('】', '')}

    # 解答部分
    answer = row['解答'].replace('\n', '、')

    # 意味部分 (●以降と最後の改行を削除)
    meaningQ = row['問題熟語の意味']
    meaningQ = meaningQ[1:]
    if '◇' in meaningQ:
        meaningQ = meaningQ.split('◇')[0]
    if '●' in meaningQ:
        meaningQ = meaningQ.split('●')[0]
    meaningQ = meaningQ.replace('┌', '').replace('├', '').replace('└', '')
    meaningQ = meaningQ.rstrip('\n').replace('\n', '<br>')

    meaningA = row['解答熟語の意味']
    meaningA = meaningA[1:]
    if '◇' in meaningA:
        meaningA = meaningA.split('◇')[0]
    if '●' in meaningA:
        meaningA = meaningA.split('●')[0]
    meaningA = meaningA.replace('┌', '').replace('├', '').replace('└', '')
    meaningA = meaningA.rstrip('\n').replace('\n', '<br>')

    meaning = meaningQ + '<br><br>' + meaningA

    # ID
    id_value = row['ID']

    # 難易度
    level_value = int(row['難易度'].replace('◆', ''))

    # 選択肢
    if '〕' in meaningA and '<br>' in meaningA:
        start_index = meaningA.index('〕') + 1
        end_index = meaningA.index('<br>')
        choice = meaningA[start_index:end_index]
        choice = jaconv.kata2hira(choice)
    else:
        choice = ""
        print(f"解答の読みがない\n{meaningA}")

    # サブ質問データを作成
    sub_question = {
        "id": id_value,
        "level": level_value,
        "questionParts": question_parts,
        "answer": answer,
        "explanation": meaning,
        "choice": choice
    }

    # サブ質問をJSONデータに追加
    json_data["subQuestions"].append(sub_question)

# JSONファイルに書き込む
with open("7-1.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=4)
```
